**NB - Regression**

```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('C:/MAFAS/APU/CT046-3-M-AML/CT046 - LABS/Python LAB MATERIALS/Lab 4 - Naive Bayes/kc_house_data.csv')
data.head()
```

```python
data.shape
```

```python
# Drop unwanted variables
data = data.drop(['id', 'date', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long'], axis = 1)
```

```python
data.columns
```

```python
# Descriptive statistics
data.describe()
```

```python
# Detect missing values
data.isnull().sum()
```

```python
# Variable assingment
x = data.drop('price', axis = 1)
y = data['price']
```

```python
y.head()
```

```python
# Data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
x_test[0:5]
```

```python
y_test.head()
```

**MultinomialNB**

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
y_pred = mnb.predict(x_test)
```

```python
# MSE, RMSE, MAE

print("Mean Squared Error: ", round(mean_squared_error(y_test, y_pred), 2))
print('Root Mean Squared Error: ', round(root_mean_squared_error(y_test, y_pred), 2))
print('Mean Absolute Error: ', round(mean_absolute_error(y_test, y_pred), 2)) 
```

**GaussianNB**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
```

```python
# MSE, RMSE, MAE

print("Mean Squared Error: ", round(mean_squared_error(y_test, y_pred), 2))
print('Root Mean Squared Error: ', round(root_mean_squared_error(y_test, y_pred), 2))
print('Mean Absolute Error: ', round(mean_absolute_error(y_test, y_pred), 2)) 
```

**NOTE:**
* Feature scaling is NOT required for GaussianNB, but in practice, it is usually a very good idea.
* For General Regression Models (like Linear, Ridge, Lasso): Normalization is strongly recommended.
* Among the MultinomialNB and GaussianNB models, GaussianNB model is more suitable for regression problems. MultinomialNB model is more suitable for text classification problems.
* GaussianNB or MultinomialNB is a classification model, not a regression model. Choose a suitable regression model.
* No Hyperparameter tuning is required for both MultinomialNB and GaussianNB as they already smoothed models. 
* Hence, a Cross-Validation can be done to further validate the models.Cross-Validation techniques will be discuss in future sessions.

**K-Fold CV**

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html#sklearn.model_selection.KFold

The performance measure reported by K-Fold cross-validation is then the average of the values computed in the loop. This approach can be computationally expensive, but does not waste too much data (as is the case when fixing an arbitrary validation set), which is a major advantage in problems such as inverse inference where the number of samples is very small.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold, cross_val_score
folds = KFold(n_splits = 5, shuffle = True, random_state = 42)
gnb = GaussianNB()
R2s_kcv = cross_val_score(gnb, x_train, y_train, cv = folds, scoring = 'neg_root_mean_squared_error')
print(R2s_kcv)
print("Average RMSE after K-Fold CV: ", np.mean(R2s_kcv))
```

**Repeated KFold**

```python
from sklearn.model_selection import RepeatedKFold
rkfolds = RepeatedKFold(n_splits = 5, n_repeats = 5, random_state = 42)

R2s_rkcv = cross_val_score(gnb, x_train, y_train, cv = rkfolds, scoring = 'neg_root_mean_squared_error')
print(R2s_rkcv)
print("Average RMSE after Repeated K-Fold CV: ", np.mean(R2s_rkcv))
```
