**Simple Linear Regression**

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
# Importing the dataset
dataset = pd.read_csv('C:/MAFAS/APU/CT046-3-M-AML/CT046 - LABS/Python LAB MATERIALS/Lab 6 - Linear Regression/kc_house_data.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[dataset.columns[5]] # sqft_living
y = dataset[dataset.columns[2]] # price
```

```python
X = sm.add_constant(x)
est = sm.OLS(y, X).fit()
print(est.summary())
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
```

```python
# predicting on X_test....This is a rare occasion to predict the TV using one IV
x_1 = dataset['sqft_living'].values.reshape(-1, 1)
y_1 = dataset['price'].values.reshape(-1, 1)

slr.fit(x_1, y_1)
y_pred = slr.predict(x_1)
y_pred.round(2)
```

```python
from sklearn.metrics import r2_score
print('R2 Score:', round(r2_score(y_1, y_pred), 2))
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared Error: ", round(mean_squared_error(y_1, y_pred), 3)) 
```

**Multiple Linear Regression**

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
# Importing the dataset
data = pd.read_csv('C:/MAFAS/APU/CT046-3-M-AML/CT046 - LABS/Python LAB MATERIALS/Lab 6 - Linear Regression/50_Startups.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['State'].describe()
```

```python
# Drop State variable
data = data.drop(['State'], axis = 1)
```

```python
data.columns
```

```python
x = data.drop(['Profit'], axis = 1)
y = data['Profit']
```

```python
# Multiple Linear Regression
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x, y)
round(mlr.score(x, y), 2) # Return the R^2 
```

```python
print(mlr.coef_)
print(round(mlr.intercept_, 3))
```

```python
X = sm.add_constant(x)
est = sm.OLS(y, X)
est2 = est.fit()
print(est2.summary())
```

```python
# Predictions
predictions = mlr.predict(x)
predictions[0:2].round(2)
```

**Build a Linear Regression predictive model**

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# training a linear regression model on train
mlr.fit(x_train, y_train)
```

```python
# R2 Score = Accuracy
# Train Accuracy
print('Train Accuracy = ', round(mlr.score(x_train, y_train) * 100, 2), '%')
# Test Accuracy....This is the one we consider when evaluating the model
print('Test Accuracy = ', round(mlr.score(x_test, y_test)* 100, 2), '%')
```

```python
# predicting on x_test
y_pred = mlr.predict(x_test)
y_pred.round(2)
```

```python
from sklearn.metrics import r2_score
round(r2_score(y_test, y_pred), 2)
```

```python
pd.DataFrame({'Actual': y_test.round(2), 'Predicted': y_pred.round(2)})
```

```python
# calculating MSE, RMSE, MAE
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("Mean Squared Error: ", round(mean_squared_error(y_test, y_pred), 2))
print('Mean Absolute Error: ', round(mean_absolute_error(y_test, y_pred), 2))  
print('Root Mean Squared Error: ', np.sqrt(mean_squared_error(y_test, y_pred)).round(2))
```
