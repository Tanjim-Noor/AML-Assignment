**Decision Tree - Multiclass**

```python
# Load libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import arange
from IPython.display import Image  
import pydotplus
from sklearn.tree import export_graphviz
from six import StringIO  
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
iris = pd.read_csv("C:/MAFAS/APU/CT046-3-M-AML/CT046 - LABS/Python LAB MATERIALS/Lab 5 - Decision Tree/iris.csv")
iris.head()
```

```python
iris.isnull().sum()
```

```python
print(iris['Species'].value_counts())
sns.countplot(x = 'Species', data = iris)
```

```python
x = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']
```

```python
y.head()
```

```python
# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
y_test.head()
```

Labelling the target variable

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
# Create decision tree classifer object
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)

y_pred = dtc.predict(x_test)

print(dtc.get_params())

# Model Accuracy
print("\nAccuracy:", round(accuracy_score(y_test, y_pred), 2))
```

```python
import matplotlib.pyplot as plt
feature_imp = pd.Series(dtc.feature_importances_, index = x_train.columns).sort_values(ascending = False)

# Creating a bar plot, displaying only the top k features
sns.barplot(x = feature_imp[:10], y = feature_imp.index[:10])
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Important Features")
plt.show()
```

**Grid Search CV**

```python
dt = DecisionTreeClassifier()

# Hyperparameter Optimization
parameters = {'max_features': ['log2', 'sqrt','auto'], 
              'criterion': ['entropy', 'gini', 'log_loss'],
              'max_depth': arange(2, 10, 1),
              'min_samples_split': arange(2, 10, 1),
              'min_samples_leaf': arange(1, 10, 1)
             }

# Run the grid search
grid_search_dt = GridSearchCV(dt, parameters, scoring = 'accuracy', cv = 7)

grid_search_dt.fit(x_train, y_train)
best_parameters_dt = grid_search_dt.best_params_  
best_score_dt = grid_search_dt.best_score_ 
print(best_parameters_dt)
print(best_score_dt)

y_pred_1 = grid_search_dt.predict(x_test)

# Get the accuracy score
dt_acc=accuracy_score(y_test, y_pred_1)*100
dt_pre=precision_score(y_test, y_pred_1, average = 'weighted')
dt_recall=recall_score(y_test, y_pred_1, average = 'weighted')
dt_f1_=f1_score(y_test, y_pred_1, average = 'weighted')

print("DT - Accuracy: {:.3f}.".format(dt_acc))
print("DT - Precision: {:.3f}.".format(dt_pre))
print("DT - Recall: {:.3f}.".format(dt_recall))
print("DT - F1_Score: {:.3f}.".format(dt_f1_))
```
